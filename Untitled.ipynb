```python
def find_nodes(txn_records_o, acct):
    new_txn_records = []
    new_acct_list = [acct]
    incre = 0
    def find(txn_records_o, new_txn_records, incre, new_acct_list):
        if incre == len(new_acct_list):
            pass
        else:
            for acct in new_acct_list:
                for i in txn_records_o:
                    if i['source'] == acct or i['target'] == acct:
                        if i not in new_txn_records:
                            new_txn_records.append(i)
                for i in new_txn_records:
                    if i['source'] not in new_acct_list:
                        new_acct_list.append(i['source'])
                for i in new_txn_records:
                    if i['target'] not in new_acct_list:
                        new_acct_list.append(i['target']) 
                new_acct_list = list(OrderedDict.fromkeys(new_acct_list))
                incre = len(new_acct_list)
            find(txn_records_o, new_txn_records, incre, new_acct_list)
        return new_acct_list, new_txn_records
    new_acct_list, new_txn_records = find(txn_records_o, new_txn_records, incre, new_acct_list)
    return new_txn_records, new_acct_list  
```
